```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Load Titanic dataset
titanic = sns.load_dataset("titanic")
```

```python

print(titanic.head(20))
```

```python
# Set Seaborn global theme for nicer style
sns.set_theme(style="whitegrid", palette="bright")
```

```python
# --------------------------------------
# 1. Scatter Plot
# --------------------------------------
# About: Scatter plots show the relationship between two continuous variables.
# Use case: Useful to find correlation, patterns, or clustering in data.

 # The dataset (Titanic) we are plotting from
    # x="age",            : x-axis  passenger's age
    # y="fare",           : y-axis  ticket fare
    # hue="survived",     : color points by survival status (0 = not survived, 1 = survived)
    # style="sex",        : change marker style (male/female different shapes)
    # s=80,               : size of each point (marker size)
    # edgecolor="black",  : black outline around each point
    # alpha=0.7           : transparency (0 = invisible, 1 = solid; 0.7 = slightly transparent)
sns.scatterplot(data=titanic, x="age", y="fare", hue="survived", style="sex", s=80, edgecolor="black", alpha=0.7)
plt.title("Age vs Fare (Survival Colored, Sex Styled)", fontsize=14, weight="bold")
plt.show()
```

```python
# --------------------------------------
# 2. Line Plot
# --------------------------------------
# About: Line plots connect data points to show trends over a variable.
# Use case: Great for analyzing trends or changes in numeric data over age, time, etc.
sns.lineplot(data=titanic, x="age", y="fare", hue="pclass", linewidth=2.2 , palette="bright")
plt.title("Fare vs Age by Class", fontsize=14, weight="bold")
plt.show()
```

```python
# --------------------------------------
# 3. Bar Plot
# --------------------------------------
# About: Bar plots show average values of a category.
# Use case: Compare mean/median values of groups (like fare across classes/sex).

# by default, plots like sns.barplot() or sns.pointplot() show 95% confidence intervals as error bars.
# They indicate how much the value (like mean fare) could vary if we sampled again.

sns.barplot(data=titanic, x="pclass", y="fare", hue="sex", errorbar=None)
plt.title("Average Fare by Class & Sex", fontsize=14, weight="bold")
plt.show()
```

```python
# --------------------------------------
# 4. Count Plot
# --------------------------------------
# About: Count plots show frequency (count) of categorical variables.
# Use case: Best for distribution of classes, e.g., passengers per class.
sns.countplot(data=titanic, x="pclass", hue="survived")
plt.title("Passenger Count by Class & Survival", fontsize=14, weight="bold")
plt.show()
```

```python
# --------------------------------------
# 5. Box Plot
# --------------------------------------
# About: Box plots display data distribution using quartiles and outliers.
# Use case: Identify spread, median, and extreme values across groups.
sns.boxplot(data=titanic, x="pclass", y="fare", hue="survived")
plt.title("Fare Distribution by Class & Survival", fontsize=14, weight="bold")
plt.show()
```

```python
# --------------------------------------
# 6. Violin Plot
# --------------------------------------
# About: Violin plots combine box plot with KDE for data distribution.
# Use case: Compare full distribution of continuous data across categories.
sns.violinplot(data=titanic, x="pclass", y="age", hue="sex", split=True)
plt.title("Age Distribution by Class & Sex", fontsize=14, weight="bold")
plt.show()
```

```python
# --------------------------------------
# 7. Strip Plot
# --------------------------------------
# About: Strip plots scatter individual data points along a category.
# Use case: Show exact distribution & spread of values with slight jitter.
sns.stripplot(data=titanic, x="pclass", y="fare", hue="survived", dodge=True, alpha=0.7, jitter=True)
plt.title("Fare by Class & Survival (Strip Plot)", fontsize=14, weight="bold")
plt.show()
```

```python
# --------------------------------------
# 8. Swarm Plot
# --------------------------------------
# About: Swarm plots arrange data points to avoid overlap.
# Use case: Similar to strip plots but clearer when points overlap.
sns.swarmplot(data=titanic, x="pclass", y="fare", hue="survived", dodge=True)
plt.title("Fare by Class & Sex (Swarm Plot)", fontsize=14, weight="bold")
plt.show()
```

```python
# --------------------------------------
# 9. Histogram
# --------------------------------------
# About: Histograms group continuous data into bins to show distribution.
# Use case: Good for frequency distribution, here age distribution by survival.
sns.histplot(data=titanic, x="age", bins=30, hue="survived", kde=True)
plt.title("Age Distribution by Survival", fontsize=14, weight="bold")
plt.show()
```

```python
# --------------------------------------
# 10. KDE Plot
# --------------------------------------
# About: KDE (Kernel Density Estimate) smooths data to show probability density.
# Use case: Compare continuous distributions smoothly (e.g., fare for survival).
sns.kdeplot(data=titanic, x="fare", hue="survived", fill=True, alpha=0.6, linewidth=2)
plt.title("Fare Distribution by Survival (KDE)", fontsize=14, weight="bold")
plt.show()
```

```python
# --------------------------------------
# 11. Pair Plot
# --------------------------------------
# About: Pair plots show pairwise relationships between multiple features.
# Use case: Great for multivariate analysis with scatter plots & histograms.
sns.pairplot(titanic[["age", "fare", "pclass", "survived"]].dropna(), hue="survived", diag_kind="kde")
plt.suptitle("Pair Plot: Titanic Features", y=1.02, fontsize=14, weight="bold")
plt.savefig("Pair_Plot_titanic.png", dpi=300, bbox_inches="tight")  
plt.show()
```

```python
# --------------------------------------
# 12. Heatmap
# --------------------------------------
# About: Heatmaps visualize matrix-like data with color intensity.
# Use case: Show correlations between numerical features in Titanic dataset.
corr = titanic.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5, linecolor="white", square=True)
plt.title("Titanic Dataset Correlation Heatmap", fontsize=14, weight="bold")
plt.savefig("Heatmap_titanic.png", dpi=300, bbox_inches="tight")  
plt.show()
```
